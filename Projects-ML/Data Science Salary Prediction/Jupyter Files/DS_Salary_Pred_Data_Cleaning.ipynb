```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('Glassdoor_jobs.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#salary parsing
df['Salary Estimate'].unique()
```

```python
#getting to know if its hourly salary or the salary is provided by employer
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary' in x.lower() else 0)
```

```python
df = df[df['Salary Estimate'] != '-1']
```

```python
salary = df['Salary Estimate'].apply(lambda x : x.split('(')[0])
```

```python
salary
```

```python
#removing k and dollars
salary_min_kd = salary.apply(lambda x: x.replace('K','').replace('$',''))
```

```python
#removing Per Hour and Employer Provided Salary:
salary_min_pe = salary_min_kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
salary_min_pe.unique()
```

```python
df['min_salary'] = salary_min_pe.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = salary_min_pe.apply(lambda x: int(x.split('-')[1]))
df['avg_salary'] = (df.min_salary+df.max_salary)/2
```

```python
#2nd company name 
df['Company Name'].str.split('\n')[0][0]
```

```python
df['Company Name'].unique()
```

```python
df['Company Name'] = df['Company Name'].apply(lambda x: x.split('\n')[0])
```

```python
df.head()
```

```python
#state field
#removing the NM etc shortforms
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
```

```python
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis = 1)
```

```python
df['Location'] = df['Location'].apply(lambda x: x.split(',')[0])
```

```python
df['Founded'].unique()
```

```python
#age of company 
df['age'] = df['Founded'].apply(lambda x: x if x <1 else 2020 - x)
```

```python
df.head()
```

```python
df['age'].unique()
```

```python
#python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
 
#r studio 
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df.R_yn.value_counts()

#spark 
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df.spark.value_counts()

#aws 
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df.aws.value_counts()

#excel
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df.excel.value_counts()
```
